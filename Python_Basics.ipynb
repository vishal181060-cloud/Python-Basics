```python
#Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.
my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

print("Original list:", my_list)
my_list[1] = 99
print("Modified list:", my_list)

print("\nOriginal tuple:", my_tuple)
try:
    my_tuple[1] = 99
except TypeError as e:
    print("Error when modifying tuple:", e)

```

```python
#Write a function to demonstrate the behavior of mutable and immutable arguments.
def modify_arguments(num, my_list):
    # num is immutable (int)
    num += 10

    # my_list is mutable (list)
    my_list.append(99)

    print("Inside function:")
    print("num =", num)
    print("my_list =", my_list)

# Main program
number = 5
numbers_list = [1, 2, 3]

print("Before function call:")
print("number =", number)
print("numbers_list =", numbers_list)

modify_arguments(number, numbers_list)

print("\nAfter function call:")
print("number =", number)          # unchanged (immutable)
print("numbers_list =", numbers_list)  # changed (mutable)
```

```python
#Write a program to reverse a string using a while loop.
s = input("Enter a string: ")
reversed_str = ""
i = len(s) - 1

while i >= 0:
    reversed_str += s[i]
    i -= 1

print("Reversed string:", reversed_str)
```

```python
#Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial of", num, "is", fact)
```
